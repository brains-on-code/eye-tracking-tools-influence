```python
import csv

# Define the input TSV file and output CSV file paths
input_file = 'bachelorthesis/P114.tsv'
output_file = 'bachelorthesis/P114_time_x_y.csv'

# Initialize a list to store the data
data_list = []

# Read the TSV file and extract the required columns
with open(input_file, 'r', newline='') as tsvfile:
    tsvreader = csv.reader(tsvfile, delimiter='\t')
    next(tsvreader)  # Skip the header row
    
    for row in tsvreader:
        timestamp = row[0]  # Assuming timestamp is in the first column
        x_cord = row[6]     # Assuming X coordinate is in the third column
        y_cord = row[7]     # Assuming Y coordinate is in the fourth column
        
        data_list.append([timestamp, x_cord, y_cord])

# Write the data to a CSV file
with open(output_file, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['Timestamp', 'X_Coord', 'Y_Coord'])  # Write the header row
    csvwriter.writerows(data_list)

print(f"CSV file saved to {output_file}")
```
